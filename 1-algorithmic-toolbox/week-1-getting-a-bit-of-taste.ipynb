# Question2(Q2): Solving the maximum pairwise product
[Problem link (geeksforgeeks)](https://www.geeksforgeeks.org/return-a-pair-with-maximum-product-in-array-of-integers/)

Given an array with both +ive and -ive integers, return a pair with highest product.
```
Input: arr[] = {1, 4, 3, 6, 7, 0}  
Output: {6,7}  

Input: arr[] = {-1, -3, -4, 2, 0, -5} 
Output: {-4,-5} 
```

## My 1st try
- Perhaps the key to this is to look for the biggest two integers because they will make the biggest product. Exceptions though:
    - 0
    - -ve integers with +ve integers
- But if all integers are -ve, the solution would be looking for the smallest two integers. 
- So I resorted to finding all the two smallest and biggest numbers -- four numbers in all.

````python
from random import sample, randint

# CONDITION: len(arr) >= 2
def maxPairwiseProduct(arr):
    
    # find the first smallest & biggest
    smallest1 = min(arr)
    biggest1 = max(arr)
    
    # find the second smallest & biggest 
    if len(arr) <= 2:
        print((smallest1, biggest1))
    else:
        smallest2 = min(list(filter( lambda x : x > smallest1 and x != smallest1, arr)))
        biggest2 = max(list(filter(lambda x : x < biggest1 and x != biggest1, arr)))
        # Just for checking
        # print(smallest1, smallest2, biggest1, biggest2)
        pair = (smallest1, smallest2) if smallest1 * smallest2 > biggest1 * biggest2 else (biggest1, biggest2)
        print(pair)
        
# Just for making random arrays 
def getRandArrs(n):
    arrContainer = []
    for _ in range(n):
        # mixed
        arrContainer.append(sample(range(-100, 100), k=randint(2,100)))
        # +ve only
        arrContainer.append(sample(range(0, 200), k=randint(2,100)))
        # -ve only
        arrContainer.append(sample(range(-200, 0), k=randint(2,100)))   
    return arrContainer

randArrs = getRandArrs(20)    
for arr in randArrs:
    maxPairwiseProduct(arr)
````

# Answer2(A2): Solving the maximum pairwise product
I looked at the answer from geeksforgeeks. And it provided **three solutions**:
1. Simple: brute-forcing by going over all values
2. Better: sort then get the product of last two numbers
**3. Efficient: keep track of 1st and 2nd negative & positive values**

Wow. Although it took some time for me, I got've got the efficient solution on the first try. This is the answer from the geeksforgeeks website:

````python
# Function to find maximum product  
# pair in arr[0..n-1] 
def maxProduct(arr, n): 
  
    if (n < 2): 
        print("No pairs exists") 
        return
  
    if (n == 2): 
        print(arr[0] ," " , arr[1]) 
        return
  
    # Iniitialize maximum and  
    # second maximum 
    posa = 0
    posb = 0
  
    # Iniitialize minimum and  
    # second minimum 
    nega = 0
    negb = 0
  
    # Traverse given array 
    for i in range(n): 
      
        # Update maximum and second 
        # maximum if needed 
        if (arr[i] > posa): 
            posb = posa 
            posa = arr[i] 
          
        elif (arr[i] > posb): 
            posb = arr[i] 
  
        # Update minimum and second  
        # minimum if needed 
        if (arr[i] < 0 and abs(arr[i]) > abs(nega)): 
            negb = nega 
            nega = arr[i] 
          
        elif(arr[i] < 0 and abs(arr[i]) > abs(negb)): 
            negb = arr[i] 
  
    if (nega * negb > posa * posb): 
        print("Max product pair is {" ,  
                nega ,", ", negb , "}") 
    else: 
        print( "Max product pair is {" , 
                 posa ,", " ,posb , "}") 
        
# Just for making random arrays 
def getRandArrs(n):
    arrContainer = []
    for _ in range(n):
        # mixed
        arrContainer.append(sample(range(-100, 100), k=randint(2,100)))
        # +ve only
        arrContainer.append(sample(range(0, 200), k=randint(2,100)))
        # -ve only
        arrContainer.append(sample(range(-200, 0), k=randint(2,100)))   
    return arrContainer

# For comparison!
def maxPairwiseProduct(arr):
    
    # find the first smallest & biggest
    smallest1 = min(arr)
    biggest1 = max(arr)
    
    # find the second smallest & biggest 
    if len(arr) <= 2:
        print((smallest1, biggest1))
    else:
        smallest2 = min(list(filter( lambda x : x > smallest1 and x != smallest1, arr)))
        biggest2 = max(list(filter(lambda x : x < biggest1 and x != biggest1, arr)))
        # Just for checking
        # print(smallest1, smallest2, biggest1, biggest2)
        pair = (smallest1, smallest2) if smallest1 * smallest2 > biggest1 * biggest2 else (biggest1, biggest2)
        print(pair)

randArrs = getRandArrs(20)    
for arr in randArrs:
    maxProduct(arr, len(arr))
    maxPairwiseProduct(arr)
````
